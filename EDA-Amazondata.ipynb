```python
# categorical 
```

```python
sns.countplot(x=df['product_category'],palette='viridis',data=df)
```

```python
#insights >>  Beauty is one of the highest count of the product_category in Amazon 
```

```python
# df['payment_method'].value_counts().plot.pie(autopct='%1.1f%%',label=lang,explode=e)
e = (0.0,0.1,0.1,0.1,0.1)
lang = df.payment_method.unique()
student = df['payment_method'].value_counts()
plt.pie(student,labels=lang,autopct="%1.1f%%",explode=e)
plt.title("Payment Method Using in Amazon")
plt.show()
```

```python
# insights >> credit card and upi is share the equal using in amazon for payment_method 
```
